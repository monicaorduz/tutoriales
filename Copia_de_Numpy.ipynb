NumPy (Numerical Python) es una biblioteca que ofrece:

- Objetos de arrays multidimensionales (`ndarray`) permite trabajar con arreglos en varias dimensiones, matem√°ticamente similar a vectores y matrices.
- Operaciones r√°pidas sobre esos arrays (matem√°ticas, estad√≠sticas, l√≥gicas, etc.)
- Funciones de √°lgebra lineal, transformadas de Fourier, simulaciones aleatorias, lectura/escritura de archivos, entre muchas otras herramientas cient√≠ficas.


##  <font color='#2184A1'>El coraz√≥n de NumPy: `ndarray`</font>

El objeto central de NumPy es el **`ndarray`**, un array (arreglo) de **n dimensiones**, que:

- Contiene **datos homog√©neos** (todos del mismo tipo)
- Tiene **tama√±o fijo** desde su creaci√≥n (no crece como las listas)
- Es procesado internamente con **c√≥digo compilado** para mayor velocidad


##  <font color='#2184A1'>Diferencias con listas de Python</font>

| Caracter√≠stica               | Listas de Python              | Arrays de NumPy (`ndarray`)       |
|-----------------------------|-------------------------------|-----------------------------------|
| Tama√±o                      | Din√°mico                      | Fijo                              |
| Tipo de datos               | Puede mezclar tipos           | Homog√©neo (misma clase de datos) |
| Velocidad                   | M√°s lentas para c√°lculos      | Mucho m√°s r√°pidas                 |
| Funcionalidad matem√°tica    | Limitada                      | Muy extensa y optimizada          |

>  Se pueden hacer arrays de objetos si se desea heterogeneidad, pero se pierde eficiencia.


##  <font color='#2184A1'>¬øPor qu√© usar NumPy?</font>

- Es **m√°s eficiente** que las estructuras nativas para datos num√©ricos
- Permite **operaciones vectorizadas**, sin bucles expl√≠citos
- Es compatible con la mayor√≠a de librer√≠as cient√≠ficas en Python (Pandas, SciPy, Scikit-learn, etc.)
- Muchas funciones esperan arrays de NumPy como entrada o salida

Al tener arreglos en NumPy, para acceder a sus elementos para analizar o modificar. Esto se hace mediante:

-  **Indexaci√≥n**: acceder a una posici√≥n espec√≠fica.
-  **Slicing**: acceder a un subconjunto del arreglo.

## <font color='#2184A1'>Indexaci√≥n en arreglos unidimensionales</font>
La indexaci√≥n es el proceso mediante el cual se accede a elementos individuales dentro de un arreglo NumPy. En el caso de arreglos unidimensionales (es decir, vectores), cada elemento tiene una posici√≥n √∫nica llamada √≠ndice, que comienza en 0 para el primer elemento.

Reglas generales:

- Los √≠ndices van desde 0 hasta n-1, donde n es el n√∫mero total de elementos en el arreglo.

- Se puede acceder a un elemento utilizando la notaci√≥n a[i], donde i es el √≠ndice deseado.

- Tambi√©n se pueden usar √≠ndices negativos:

  - -1 se refiere al √∫ltimo elemento

  - -2 al pen√∫ltimo, y as√≠ sucesivamente

```python
import numpy as np

a = np.array([10, 20, 30, 40, 50])

print("Elemento en la posici√≥n 0:", a[0]) # Muestra 10
print("Elemento en la posici√≥n 3:", a[3]) # Muestra 40
print("√öltimo elemento:", a[-1]) # Muestra 50
```
##  <font color='#2184A1'>Slicing en arreglos unidimensionales</font>
El slicing (o segmentaci√≥n) es una t√©cnica que permite acceder a un subconjunto de elementos contiguos dentro de un arreglo. En NumPy, se utiliza la notaci√≥n:
$$a[inicio:fin:paso]$$
donde $a$ es un arreglo.

**Descripci√≥n de los par√°metros:**

- inicio: √≠ndice donde comienza el corte (incluido). Si se omite, se asume 0.

- fin: √≠ndice donde termina el corte (excluido). Si se omite, se asume el final del arreglo.

- paso: cada cu√°ntos elementos se toma uno. Si se omite, se asume paso 1.

Tambi√©n es posible usar √≠ndices negativos para contar desde el final del arreglo.


````python
a=np.random.randint(1,20,10)
print(f'Imprime un arreglo de 10 valores entre 1 y 20 : {a}')

print(f'Imprime los valores en posiciones pares de los primeros cinco elementos de $a$ :{a[0:5:2]}')

print(f'Imprime los valores en posiciones impares de los √∫ltimos cinco elementos de $a$ :{a[-1:-6:-2]}')
````

## <font color='#2184A1'>Indexaci√≥n en arreglos bidimensionales</font>
En los arreglos de dos dimensiones (matrices), los elementos est√°n organizados por filas y columnas. Para acceder a un elemento espec√≠fico se utiliza la notaci√≥n:

$$a[i,j]$$
donde:

- $i$ es el √≠ndice de la fila (comenzando en 0)

- $j$ es el √≠ndice de la columna (comenzando en 0)

Tambi√©n es posible:

- Acceder a toda una fila usando $a[i]$ o $a[i, :]$

- Acceder a toda una columna usando $a[:, j]$

Usar √≠ndices negativos para acceder desde el final hacia el principio

```python
import numpy as np

A = np.array([[10, 20, 30],
              [40, 50, 60],
              [70, 80, 90]])

print(A)

# Acceder al elemento en la fila 1, columna 2 (recordemos que se empieza en 0)
print("\nElemento A[1, 2] (fila 1, columna 2):", A[1, 2])  # Muestra 60

# Acceder a toda la fila 0
print("\nFila 0 completa (A[0]):", A[0])  # Muestra [10 20 30]

# Acceder a toda la fila 2 con notaci√≥n expl√≠cita
print("Fila 2 completa (A[2, :]):", A[2, :])  # Muestra [70 80 90]

# Acceder a toda la columna 1
print("Columna 1 completa (A[:, 1]):", A[:, 1])  # Muestra [20 50 80]

# Acceder al √∫ltimo elemento usando √≠ndices negativos
print("√öltimo elemento (A[-1, -1]):", A[-1, -1])  # Muestra 90
```
 Slicing en arreglos bidimensionales
El slicing en matrices permite extraer submatrices a partir de un conjunto de filas y columnas.

Se utiliza la notaci√≥n:

$$
a[i_1:i_2,\ j_1:j_2]
$$

donde:

- `i_1:i_2` define el **rango de filas**, desde la fila \\( i_1 \\) (incluida) hasta \\( i_2 \\) (excluida).
- `j_1:j_2` define el **rango de columnas**, desde la columna \\( j_1 \\) (incluida) hasta \\( j_2 \\) (excluida).

Tambi√©n es v√°lido usar un **paso** (stride) como tercer argumento en cada dimensi√≥n:

$$
a[i_1:i_2:s_i,\ j_1:j_2:s_j]
$$

donde \\( s_i \\) y \\( s_j \\) son los pasos en filas y columnas respectivamente.

Adem√°s, se pueden usar **√≠ndices negativos** para contar desde el final del arreglo.

````python
import numpy as np


A = np.arange(16).reshape(4, 4)  # Crea una matriz 4x4 con valores de 0 a 15
print(f'Matriz {A}') # Muestra la matriz A

print(f'Submatriz $2\times 2$ superior {A[0:2, 0:2]}') #imprime la submatriz 2x2 desde la esquina superior izquierda
print(f'Submatriz central {A[1:3, 1:3]}') # Imprime la sumatriz central
print(f'Submatriz √∫ltimas dos filas {A[2:, :]}') # Muestra las √∫ltimas dos filas
print(f'Submatriz columnas pares de dos en dos {A[:, ::2]}') # Muestra las columnas pares de dos en dos
print(f'submatriz con paso en ambas direcciones {A[::2, ::2]}') # Muestra submatriz con paso en ambas direcciones
````
