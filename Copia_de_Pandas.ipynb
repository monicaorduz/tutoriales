Una de las operaciones m√°s potentes en pandas es la **agrupaci√≥n**, que permite agrupar filas que comparten un mismo valor en una o m√°s columnas y luego aplicar una operaci√≥n como suma, promedio, conteo, etc.

Esto es esencial para **resumir grandes vol√∫menes de datos** y generar estad√≠sticas por categor√≠a, como EPS, tipo de sanci√≥n, o a√±o.

### <font color='#2184A1'> Funciones clave</font>

| Funci√≥n | ¬øPara qu√© sirve? |
|--------|-------------------|
| `df.groupby("col")["otra_col"].sum()` | Agrupa por una categor√≠a y suma |
| `df.groupby("col").mean()` | Calcula promedio por grupo |
| `df.groupby("col").count()` | Cuenta elementos por grupo |
| `df.groupby(["col1", "col2"])` | Agrupa por m√∫ltiples columnas |
| `df["col"].value_counts()` | Cuenta frecuencia de cada valor en una columna |

> Estas funciones permiten crear reportes agregados por EPS, a√±o, tipo de sanci√≥n, etc.

## <font color='#2184A1'>Manejo de fechas</font>

En muchos an√°lisis de datos es importante trabajar correctamente con **fechas**. Esto permite ordenar cronol√≥gicamente, agrupar por a√±os o meses, y realizar comparaciones temporales.  
En `pandas`, las fechas deben estar en formato `datetime` para acceder a atributos como a√±o, mes, d√≠a o incluso el d√≠a de la semana.

### <font color='#2184A1'>Funciones clave</font>

| Funci√≥n | ¬øPara qu√© sirve? |
|--------|-------------------|
| `pd.to_datetime(df["col"])` | Convierte texto a tipo fecha |
| `df["col"].dt.year` | Extrae el a√±o de una fecha |
| `df["col"].dt.month` | Extrae el n√∫mero de mes |
| `df["col"].dt.day` | Extrae el d√≠a del mes |
| `df.sort_values("col")` | Ordena por columna de fechas |
| `df["col"].min()` / `.max()` | Fecha m√°s antigua / reciente |

Ordenar datos es una tarea fundamental para resaltar los casos m√°s cr√≠ticos, analizar tendencias o preparar reportes.  
En `pandas`, se puede ordenar por una o m√°s columnas, en orden ascendente o descendente, y tambi√©n es posible asignar **rangos** o posiciones relativas a cada valor.

Esto permite, por ejemplo, identificar las **EPS m√°s sancionadas**, los **mayores valores de sanci√≥n**, o generar tablas de posici√≥n (rankings).


### <font color='#2184A1'>Funciones clave</font>

| Funci√≥n | ¬øPara qu√© sirve? |
|--------|-------------------|
| `df.sort_values("col")` | Ordena por una columna (ascendente por defecto) |
| `df.sort_values("col", ascending=False)` | Ordena en orden descendente |
| `df.sort_values(["col1", "col2"])` | Ordena por m√∫ltiples columnas |
| `df["col"].rank()` | Asigna rangos a los valores de una columna |
| `df.nlargest(n, "col")` / `df.nsmallest(n, "col")` | Muestra los `n` valores m√°s grandes o peque√±os |
