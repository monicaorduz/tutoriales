### <font color='#2184A1'>Listas</font>

Para definir una lista en Python, se hace entre corchetes, la lista puede tener cualquier clase de objeto hasta lista incluida en ella, por ejemplo,
$$ \text{mi_lista=[2,3.5,'Gabriel',[3,4,5], 'Diana']}$$ es una lista que contiene, elementos del tipo entero, decimal, cadena y lista.

````python
mi_lista=[2,3.5,'Gabriel',[3,4,5], 'Diana']
print(mi_lista)
````

Los ndices o las localizaciones de los elementos de una lista con $n$ elementos, empiezan el 0 y va hasta $n-1$, es decir, en la tabla se muestra el indice y el elemento que corresponde a dicho ndice

|ndice| elemento|
|---|---|
|0| 2|
|1| 3.5|
|2| 'Gabriel'|
|3| [3,4,5]|
|4| 'Diana'|

Si el lector requiere ver uno o varios elementos especficos en la lista, lo puede hacer usando

```
Python

mi_lista[0]  # Muestra el elemento con ndice 0
mi_lista[3]  # Muestra elemento en la posicin 3, comenzando desde cero, en este caso es la lista [3,4,5]
mi_lista[0:3] # Muestra los elementos desde el cero hasta la posicin 2, el va hasta la la posicin 3-1=2. Tambien puede ser de la forma mi_lista[:3].
mi_lista[2:5] # Muestra los elementos desde la posicin 2 hasta el ltimo, esto tambin puede ser mi_lista[2:].
mi_lista[2:4] # Muestra los elemntos desde el ndice 2 hasta el 3.
```

````python
print(mi_lista[0]) # Muestra el elemento con ndice 0
print(mi_lista[3]) # Muestra elemento con ndice 3
print(mi_lista[0:3]) # Muestra elemento con ndice 0 hasta el elemento con ndice 2
print(mi_lista[:3])  # Muestra elemento con ndice 0 hasta el elemento con ndice 2 (otra forma)
print(mi_lista[2:5]) # Muestra elemento con ndice 2 hasta el ltimo
print(mi_lista[2:]) ## Muestra elemento con ndice 2 hasta el ltimo (otra forma)
print(mi_lista[2:4]) # Muestra elemento con ndice 2 hasta el elemento con ndice 3
````

A continuacin se describen algunos de los mtodos ms usados para manipular listas en Python. Todos ellos se aplican con la sintaxis `mi_lista.metodo(...)`.

### `mi_lista.append(x)`
Agrega el elemento `x` al final de la lista.

### <font color='#2184A1'>Tuplas</font>

Una **tupla** es una coleccin ordenada e inmutable de elementos.
Esto significa que:

- Los elementos tienen un orden definido y se pueden acceder por ndices, como en las listas.

- Una vez creada, no se puede modificar (no se puede agregar, eliminar ni cambiar elementos).

```
mi_tupla = (10, "Hola", 3.14, True)
```

Tambin se puede crear sin parntesis, usando solo comas:

```
mi_tupla = 10, "Hola", 3.14, True
```

Para tuplas con un solo elemento, se debe usar una coma al final:

```
mi_tupla=(3,)
```

Las tuplas son usadas para:

- Para proteger datos que no deben cambiarse.
- Para retornar mltiples valores desde una funcin.
- Para usar como claves en diccionarios, ya que son inmutables.
- Para representar registros fijos (por ejemplo, coordenadas, colores RGB, fechas, etc.).

###<font color='#2184A1'>Diccionarios</font>

Un **diccionario** es una coleccin no ordenada y modificable de elementos que se almacenan en pares *clave:valor*.

- Cada elemento tiene una clave nica asociada a un valor.
- Puedes acceder al valor usando su clave, no su posicin.
- Los diccionarios en Python se escriben entre llaves {}.

```
persona = {
    "nombre": "Gabriel",
    "edad": 30,
    "activo": True
}
```

- "nombre", "edad" y "activo" son claves.

- "Gabriel", 30 y True son los valores asociados.

Si se quiere acceder algn elemento, se pone la clave y el da el valor de esa clave

```
print(persona['nombre']) # Muestra 'Gabriel'
```

###<font color='#2184A1'>Mtodos tiles de diccionarios en Python</font>

A continuacin se describen algunos de los mtodos ms usados para manipular diccionarios en Python. Todos ellos se aplican con la sintaxis `mi_dicc.metodo(...)`.

Muestra todas las claves del diccionario

### `mi_dicc.values()`
Muestra todos los valores del diccionario

### `mi_dicc.items()`
Muestra todos los pares clave:valor

Devuelve el valor asociado a una clave.

Si la clave no existe, devuelve valor_por_defecto (o None si no se indica).

Modifica el valor en la clave por x.

Agrega un nuevo par *clave:valor*

### `del mi_dicc[clave]`
Elimina la clave

## <font color='#2184A1'>Control de flujo en Python</font>

El **control de flujo** permite que un programa tome decisiones y repita tareas segn ciertas condiciones.  
Aqu aprenders a usar condicionales, bucles y otras sentencias que cambian el comportamiento de tu cdigo.


###<font color='#2184A1'>Condicionales</font>

### `if`, `elif`, `else`

Permiten ejecutar cdigo solo si se cumple una condicin.

```python
Peso = 100

if Peso >= 80:
    print("Etapa 3")
elif Peso >= 50:
    print("Etapa 2")
elif Peso >= 30:
    print("Etapa 1")
else:
    print("Neonato")

```
- El `if` inicia una estructura condicional. Se ejecuta solo si la condicin que evala es verdadera.

- El `elif` (else if) permite verificar condiciones adicionales si el if no se cumple. Puedes usar tantos elif como necesites.

- El `else` se ejecuta solo si ninguna de las condiciones anteriores se cumpli. No lleva condicin.

### <font color='#2184A1'>Bucles</font>
### `for`
Se usa para repetir una accin sobre un conjunto de elementos.

```
frutas = ["manzana", "pera", "banano"]

for fruta in frutas:
    print("Me gusta la", fruta)

```
Tambin se puede usar con range() para repetir un nmero especfico de veces:

```python
for i in range(5):
    square=i**2
    print(f"el cuadrado de {i} es {square}")
```
### `while`
Se ejecuta mientras una condicin sea verdadera.

```python
contador = 0

while contador < 3:
    print("Contador:", contador)
    contador += 1
```
Asegurar de que la condicin se vuelva falsa en algn momento para evitar bucles infinitos.

###<font color='#2184A1'>Sentencias especiales</font>

### `continue`

### <font color='#2184A1'>Operadores relacionales</font>
Se usan para comparar valores:

|Operador	|Significado	|Ejemplo (x = 5)
|---|---|---|
|==|	Igual|	x == 5  True|
|!=|	Distinto|	x != 3  True|
|>|	Mayor que|	x > 2  True|
|<|	Menor que|	x < 10  True|
|>=|	Mayor o igual que|	x >= 5  True|
|<=|	Menor o igual que|	x <= 4  False|

### <font color='#2184A1'>Operadores lgicos</font>
Permiten combinar condiciones:

|Operador|	Significado|	Ejemplo|
|---|---|---|
|and|	Ambas condiciones verdaderas|	x > 3 and x < 10|
|or|	Al menos una verdadera|	x < 3 or x > 7|
|not|	Niega la condicin|	not(x > 5)|
